```python
from transformers import pipeline
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
qa = pipeline("question-answering",  model="distilbert/distilbert-base-cased-distilled-squad",
              token=os.getenv("HF_API_KEY"))
```

```python
ctx = "Today I made a Maharashtrian food item namely, Wada Paav of potato."
q = "Of what did I make a Wada Paav?"
qa(context = ctx, question=q)
```

`gradio` App

```python
import gradio as gr

def answer(question , file):

    with open(file, 'r') as readfile:
        ctx = readfile.read().rstrip()

    resp = qa(context = ctx, question=question)
    return resp['answer']

demo = gr.Interface(
    fn=answer,
    inputs=["text",gr.File(label="Information file") ],
    outputs=["text"]
)

demo.launch()
```
