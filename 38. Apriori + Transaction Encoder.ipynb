```python
import pandas as pd
import matplotlib.pylab as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
import ipywidgets as widgets
import os 
os.chdir("D:/Training/Academy/ML(Python)/Datasets")
```

```python
groceries = []
with open("groceries.csv","r") as f:groceries = f.read()
groceries = groceries.split("\n")
groceries_list = []
for i in groceries:
    groceries_list.append(i.split(","))
groceries_list[:5]
```

```python
te = TransactionEncoder().set_output(transform='pandas')
fp_df = te.fit(groceries_list).transform(groceries_list)
def generate_rules(supp, conf):
    itemsets = apriori(fp_df, min_support=supp,  use_colnames=True)
    rules = association_rules(itemsets, metric='confidence',  min_threshold=conf)
    rules = rules[['antecedents','consequents','support', 'confidence','lift']]
    return rules.sort_values(['lift','confidence'], ascending=[False, False])

widgets.interact(generate_rules, supp=(0.001, 0.5, 0.001), conf=(0.5, 1, 0.01))
```
