```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="gsarti/it5-base-news-summarization")
```

```python
news = "The court said there was no proof of secret meetings or of Abhinav Bharat being used for the activity, noting that the entire case rested on that unproven claim. It also said the voice samples presented were not up to the mark, and agencies could not establish links to the conspiracy."
```

```python
summarizer(news, max_length=250)
```
