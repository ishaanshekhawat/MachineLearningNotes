#### Boston

```python
boston = pd.read_csv("Boston.csv")
X = boston.drop('medv', axis=1)
y = boston['medv']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False).set_output(transform='pandas')
X_trn_poly = poly.fit_transform(X_train) # train set transformed into polynomial 
X_tst_poly = poly.transform(X_test) # test set transformed into polynomial 
lr = LinearRegression()
lr.fit(X_trn_poly, y_train)
y_pred = lr.predict(X_tst_poly)
mean_absolute_error(y_test, y_pred )
```

```python
poly = PolynomialFeatures(degree=3, include_bias=False).set_output(transform='pandas')
X_trn_poly = poly.fit_transform(X_train)
X_tst_poly = poly.transform(X_test)
lr = LinearRegression()
lr.fit(X_trn_poly, y_train)
y_pred = lr.predict(X_tst_poly)
mean_absolute_error(y_test, y_pred )
```

```python
poly = PolynomialFeatures(degree=4, include_bias=False).set_output(transform='pandas')
X_trn_poly = poly.fit_transform(X_train)
X_tst_poly = poly.transform(X_test)
lr = LinearRegression()
lr.fit(X_trn_poly, y_train)
y_pred = lr.predict(X_tst_poly)
mean_absolute_error(y_test, y_pred )
```

#### Housing

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
housing = pd.read_csv("Housing.csv")
```

```python
cat_hous = housing[ housing.columns[ housing.dtypes==object ] ]
num_hous = housing[ housing.columns[ housing.dtypes!=object ] ]
ohe = OneHotEncoder(drop='first', sparse_output=False).set_output(transform='pandas')
ohe_hous = ohe.fit_transform(cat_hous)
hous = pd.concat([num_hous, ohe_hous], axis=1)
X, y = hous.drop('price', axis=1), hous['price']
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False).set_output(transform='pandas')
col_trnf = ColumnTransformer([('OHE',ohe, list( housing.columns[ housing.dtypes==object ]))],
                             remainder='passthrough',
                             verbose_feature_names_out=False)
col_trnf = col_trnf.set_output(transform='pandas')
hous = col_trnf.fit_transform(housing)
```

#### Exp_Salaries

```python
from sklearn.compose import make_column_selector
```

```python
sals = pd.read_csv("Exp_Salaries.csv")
X, y = sals.drop('Salary', axis=1), sals['Salary']
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False).set_output(transform='pandas')
col_trnf = ColumnTransformer([('OHE',ohe, make_column_selector(dtype_include=object) )],
                             remainder='passthrough',
                             verbose_feature_names_out=False)
col_trnf = col_trnf.set_output(transform='pandas')
X = col_trnf.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
degrees = [1,2,3,4]
scores = []
for d in degrees:
    poly = PolynomialFeatures(degree=d, include_bias=False).set_output(transform='pandas')
    X_trn_poly = poly.fit_transform(X_train)
    X_tst_poly = poly.transform(X_test)
    lr = LinearRegression()
    lr.fit(X_trn_poly, y_train)
    y_pred = lr.predict(X_tst_poly)
    scores.append([d, mean_absolute_error(y_test, y_pred ) ])
df_scores = pd.DataFrame(scores, columns=['degree', 'score'])
df_scores.sort_values('score')
```
