```python
from transformers import pipeline
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
recognize = pipeline("token-classification", model="dslim/bert-base-NER",
                     token=os.getenv("HF_API_KEY"))
```

```python
recognize("Christopher Nolan criticized for filming The Odyssey in the troubled region of Western Sahara: 'Stop filming'")
```

```python
recognize("Oil deal, Nobel nomination and more… Is Trump moving closer to Pakistan and further away from India?")
```
