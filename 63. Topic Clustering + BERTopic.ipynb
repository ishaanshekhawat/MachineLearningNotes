```python
import pandas as pd
from bertopic import BERTopic
import numpy as np
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
```

```python
df = pd.read_csv("D:/Training/Academy/Gen AI/Cases_AI/indian_news_articles/historic_articles.csv")
df.columns
```

```python
df['Header-info'] = df['header'].fillna("") + df['subheader'].fillna("")
df['Header-info'].isnull().sum()
```

```python
topic_model = BERTopic()
topics, probs = topic_model.fit_transform(df['Header-info'].to_list())
```

```python
topic_info = topic_model.get_topic_info()
topic_info
```

```python
list( topic_info[topic_info['Topic']==9]['Representative_Docs'] )
```

```python
len( topics ), df.shape
```

```python
df['Topic'] = topics
df = df.merge(topic_info, on='Topic', how='left')
```

```python
import ipywidgets as widgets
def retreive_docs(topic):
    return df[df['Name']==topic]['Header-info']
widgets.interact(retreive_docs, topic=list(topic_info['Name'].unique()))
#retreive_docs('0_entertainment_and_is_the')
```
