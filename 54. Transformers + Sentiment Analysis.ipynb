```python
import os
os.chdir("D:/Training/Academy/ML(Python)/Datasets")
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from transformers import pipeline
from tqdm import tqdm
from transformers.utils import logging
```

```python
classify = pipeline("text-classification",
                 model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

```python
classify("I liked the book")
```

```python
result = classify("I did not like that web series")
result[0]['label']
```

```python
dataset = pd.read_csv('Restaurant_Reviews.tsv', sep = '\t')
dataset.head()
```

```python
all_reviews = dataset['Review'].to_list()
result = classify( all_reviews[0] )
if result[0]['label']=='POSITIVE':
    print(1)
else:
    print(0)
```

```python
y_pred = []
for r in tqdm(all_reviews):
    result = classify( r )
    if result[0]['label']=='POSITIVE':
        y_pred.append(1)
    else:
        y_pred.append(0) 
```

```python
y = dataset['Liked']
accuracy_score( y, y_pred )
```

```python
results = classify( all_reviews )
y_pred = []
for r in results:
    if r['label']=='POSITIVE':
        y_pred.append(1)
    else:
        y_pred.append(0)   
y = dataset['Liked']
accuracy_score( y, y_pred ) 
```
