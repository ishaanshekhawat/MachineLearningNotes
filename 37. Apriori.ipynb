```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import ipywidgets as widgets
import os 
os.chdir("D:/Training/Academy/ML(Python)/Datasets")
```

```python
fp_df = pd.read_csv('Faceplate.csv',index_col=0)
fp_df = fp_df.astype(bool)
itemsets = apriori(fp_df, min_support=0.2,  use_colnames=True)
rules = association_rules(itemsets, metric='confidence',  min_threshold=0.6)
rules = rules[['antecedents','consequents','support', 'confidence','lift']]
rules.sort_values(['lift','confidence'], ascending=[False, False])
```

```python
fp_df = pd.read_csv('Cosmetics.csv',index_col=0)
fp_df = fp_df.astype(bool)
itemsets = apriori(fp_df, min_support=0.1,  use_colnames=True)
rules = association_rules(itemsets, metric='confidence',  min_threshold=0.7)
rules = rules[['antecedents','consequents','support', 'confidence','lift']]
rules.sort_values(['lift','confidence'], ascending=[False, False])
```

```python
def generate_rules(supp, conf):
    itemsets = apriori(fp_df, min_support=supp,  use_colnames=True)
    rules = association_rules(itemsets, metric='confidence',  min_threshold=conf)
    rules = rules[['antecedents','consequents','support', 'confidence','lift']]
    return rules.sort_values(['lift','confidence'], ascending=[False, False])

widgets.interact(generate_rules, supp=(0.001, 0.5, 0.01), conf=(0.5, 1, 0.01))
```
