```python
from transformers import pipeline
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
classify = pipeline("zero-shot-classification", model="facebook/bart-large-mnli",
                token=os.getenv("HF_API_KEY"))
```

```python
news = ["'Don't care what...': Trump criticises India, Russia as 'dead economies', issues BIG statement after 25% tariff announcement",
        "Manchester United on roll in Premier League US friendly with win over Bournemouth",
        "Tamil Nadu teen dies of consuming only juice for 3 months: Experts weigh in on the fad diet",
        "Justin Trudeau spotted at Katy Perry's Montreal show, romance rumours intensify"]
topics = ['Politics','Sports','Entertainment','Life Style']
```

```python
classify(news, topics)
```

### Car

```python
problem = "My car's front light are not working."
division = ['Lights','Accesories','Engine','Gear','Chassis','Insurance']
classify(problem,division)
```

```python
problem = "When is my due date for paying the insurance premium?"
division = ['Lights','Accesories','Engine','Gear','Chassis','Insurance']
classify(problem,division)
```

### Home Appliances

```python
complaints = ["My equipment is not making my room sufficiently cool",
        "washing of clothes getting problem",
        "compressor need to be replaced. Then only cooling can happen",
        "dishes not getting cleaned"]
equipments = ["Dish Washer", "Washing Machine", "Air Conditioner", "Refrigerator"]
```
